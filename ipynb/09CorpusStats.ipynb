```python
from collections import Counter
from typing import Iterator
import matplotlib.pyplot as plt
import smart_open
```

```python
%matplotlib inline
```

```python
def get_line(filepath: str) -> Iterator[str]:
    with smart_open.smart_open(filepath) as f:
        for line in f:
            yield line.decode('utf-8').strip()
```

```python
# File Strings
TOKENSFILE = 's3://pto-us-data/text-data/titleabstract_tokens.txt'
```

```python
word_freq = Counter()
```

```python
for doc in get_line(TOKENSFILE):
    word_freq.update(doc.split())
```

```python
word_freq.most_common(100)
```

```python
len(word_freq)
```

```python
word_freq.most_common()[10000:10100]
```

```python
word_freq.most_common()[-500:-1]
```

```python
word_freq['electrolytic']
```

```python
rank_freq = {i + 1: v for i, (k, v) in enumerate(word_freq.most_common())}
```

```python
len(rank_freq)
```

```python
lt = [(k, v) for k, v in rank_freq.items()]
```

```python
rank, frequency = list(zip(*lt))
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.semilogy(rank, frequency)
ax.set_xlabel('Rank')
_ = ax.set_ylabel('Frequency')
```
